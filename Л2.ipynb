## Диаграммы рассеяния



Страны с самыми низкими и самыми высокими показателями детской смертности



```python
lowest_child_mort_country = df.loc[df['child_mort'].idxmin(), 'country']
highest_child_mort_country = df.loc[df['child_mort'].idxmax(), 'country']

print(f"Минимальное значение: {lowest_child_mort_country}")
print(f"Максимальное значение: {highest_child_mort_country}")
```

Диаграмма рассеяния: ВВП на душу населения и детская смертность



```python
df['highlight_child_mort'] = 'Other Countries'
df.loc[df['country'] == lowest_child_mort_country, 'highlight_child_mort'] = 'Lowest Child Mortality'
df.loc[df['country'] == highest_child_mort_country, 'highlight_child_mort'] = 'Highest Child Mortality'

fig = px.scatter(df, x='gdpp', y='child_mort', color='highlight_child_mort',
                 trendline='ols', title='ВВП на душу населения vs. Детская смертность с линией регрессии',
                 labels={'gdpp': 'ВВП на душу населения', 'child_mort': 'Детская смертность'},
                 hover_name='country')
fig.show()
```

Страны с самыми низкими и самыми высокими показателями ожидаемой продолжительности жизни



```python
lowest_life_expec_country = df.loc[df['life_expec'].idxmin(), 'country']
highest_life_expec_country = df.loc[df['life_expec'].idxmax(), 'country']

print(f"Минимальное значение: {lowest_life_expec_country}")
print(f"Максимальное значение: {highest_life_expec_country}")
```

Диаграмма рассеяния: ВВП на душу населения и Ожидаемая продолжительность жизни

```python
df['highlight_life_expec'] = 'Other Countries'
df.loc[df['country'] == lowest_life_expec_country, 'highlight_life_expec'] = 'Lowest Life Expectancy'
df.loc[df['country'] == highest_life_expec_country, 'highlight_life_expec'] = 'Highest Life Expectancy'

fig = px.scatter(df, x='gdpp', y='life_expec', color='highlight_life_expec',
                 trendline='ols', title='ВВП на душу населения vs. Ожидаемая продолжительность жизни с линией регрессии',
                 labels={'gdpp': 'ВВП на душу населения', 'life_expec': 'Ожидаемая продолжительность жизни'},
                 hover_name='country')
fig.show()
```

Диаграмма рассеяния: Расходы на здравоохранение и детская смертность



```python
fig = px.scatter(df, x='health', y='child_mort',
                 trendline='ols', title='Расходы на здравоохранение vs. Детская смертность с линией регрессии',
                 labels={'health': 'Расходы на здравоохранение', 'child_mort': 'Детская смертность'},
                 hover_name='country')
fig.show()
```

Диаграмма рассеяния: Расходы на здравоохранение и ожидаемая продолжительность жизни



```python
fig = px.scatter(df, x='health', y='life_expec',
                 trendline='ols', title='Расходы на здравоохранение vs. Ожидаемая продолжительность жизни с линией регрессии',
                 labels={'health': 'Расходы на здравоохранение', 'life_expec': 'Ожидаемая продолжительность жизни'},
                 hover_name='country')
fig.show()
```

Группировка стран по уровню дохода




```python
bins = [0, 3355, 9960, 22800, df['income'].max() + 1]
labels = ['Low Income', 'Lower-Middle Income', 'Upper-Middle Income', 'High Income']
df['income_group'] = pd.cut(df['income'], bins=bins, labels=labels, right=False)

print(df[['country', 'income', 'income_group']].head())
```

Визуализация распределения ВВП на душу населения по группам дохода


```python
fig = px.box(df, x='income_group', y='gdpp',
                 title='Распределение ВВП на душу населения по группам дохода',
                 labels={'income_group': 'Группа дохода', 'gdpp': 'ВВП на душу населения'})
fig.show()
```

Визуализация распределения расходов на здравоохранение по группам дохода


```python
fig = px.box(df, x='income_group', y='health',
                 title='Распределение расходов на здравоохранение по группам дохода',
                 labels={'income_group': 'Группа дохода', 'health': 'Расходы на здравоохранение (% от ВВП)'})
fig.show()
```
